```python
import pandas as pd
import ast
import matplotlib.pyplot as plt

# File paths for both models
file_path_gemma2b = r'C:\Users\Public\SLM-graph-extractor-1\Plots\gemma2b\loss.txt'
file_path_gemma7b = r'C:\Users\Public\SLM-graph-extractor-1\Plots\gemma7b\loss1.txt'

# Load and parse Gemma2b data
with open(file_path_gemma2b, 'r') as file:
    lines_gemma2b = file.readlines()
data_gemma2b = [ast.literal_eval(line.strip()) for line in lines_gemma2b]
df_gemma2b = pd.DataFrame(data_gemma2b)

# Load and parse Gemma7b data
with open(file_path_gemma7b, 'r') as file:
    lines_gemma7b = file.readlines()
data_gemma7b = [ast.literal_eval(line.strip()) for line in lines_gemma7b]
df_gemma7b = pd.DataFrame(data_gemma7b)

# Debug: Print first few rows of both DataFrames to check content
print("Gemma2b Data (First 5 Rows):")
print(df_gemma2b.head())
print("\nGemma7b Data (First 5 Rows):")
print(df_gemma7b.head())

# Plot both loss curves
plt.figure(figsize=(12, 6))
plt.plot(df_gemma2b.index, df_gemma2b['loss'], marker='', linestyle='-', label='Loss (Gemma2b)', color='blue')
plt.plot(df_gemma7b.index, df_gemma7b['loss'], marker='', linestyle='-', label='Loss (Gemma7b)', color='orange')
plt.title('Loss Curves for Gemma Models', fontsize=16)
plt.xlabel('Iteration', fontsize=14)
plt.ylabel('Loss', fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
